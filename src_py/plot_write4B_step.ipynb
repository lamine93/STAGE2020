```python
import os 
from glob import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH = "../src/data_write4B.step"
EXT = "*.csv"
```

```python
#load data
all_csv_files = [file
                 for path, subdir, files in os.walk(PATH)
                 for file in glob(os.path.join(path, EXT))]
```

```python
all_csv_files
```

```python
#read csv data 
pd.options.display.float_format = '{:.6f}'.format
df_write = [pd.read_csv(file, delimiter=" ") for file in all_csv_files]
```

```python
df_write[1].head()
```

```python
df_write[0].columns
```

```python
#donn√©es des ecritures sequentielles par un pas de 4 bytes
begin_time_seq4 = df_write[3]["temps"][0]
x_seq4 = df_write[3]["temps"] - begin_time_seq4
y_seq4 = df_write[3]["latence"]
y_seq4.head()
```

```python
#plot des donn√©es des √©critures sequentielles par un pas de 4 bytes
plt.figure(figsize=(12, 3))
plt.plot(x_seq4[:500], y_seq4[:500], "-o",linewidth =2.5)
plt.xlabel("date d'arriv√©e(s)")
plt.ylabel("latence(s)")
plt.title('√©criture de 4 Bytes par un pas de 4 Bytes')
plt.savefig("../outplot/write4B/write_step4.png")
plt.show()
```
