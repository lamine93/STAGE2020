```python
import os 
from glob import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH = "../src/data_write_rand"
EXT = "*.csv"
```

```python
all_csv_files = [file
                 for path, subdir, files in os.walk(PATH)
                 for file in glob(os.path.join(path, EXT))]
```

```python
all_csv_files
```

```python
pd.options.display.float_format = '{:.6f}'.format
df_write = [pd.read_csv(file, delimiter=" ") for file in all_csv_files]
```

```python
# 0 seq  1 rand
df_write[0].head()
```

```python
df_write[0].columns
```

```python
#données des lectures aléatoire 4 bytes
begin_time_rand4 = df_write[2]["temps"][0]
x_rand4 = df_write[2]["temps"] - begin_time_rand4
y_rand4 = df_write[2]["latence"]
y_rand4.head()
```

```python
df_write[2]["latence"].describe()
```

```python
#plot données des lectures aléatoires 4 bytes
plt.figure(figsize=(12, 3))
plt.plot(x_rand4[0:500], y_rand4[0:500], "-o",linewidth =2.5)
plt.xlabel("date d'arrivée(s)")
plt.ylabel("latence(s)")
plt.title('ecriture aléatoire de 4 bytes')
plt.savefig("../outplot/write/rand/write_rand4.png")
plt.show()
```
