```python
import os 
from glob import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH = "../src/data_read16B.step"
EXT = "*.csv"
```

```python
#load data
all_csv_files = [file
                 for path, subdir, files in os.walk(PATH)
                 for file in glob(os.path.join(path, EXT))]
```

```python
all_csv_files
```

```python
#read csv data 
pd.options.display.float_format = '{:.6f}'.format
df_read = [pd.read_csv(file, delimiter=" ") for file in all_csv_files]
```

```python
df_read[2].head()
```

```python
df_read[0].columns
```

```python
#donn√©es des lectures seq par un pas de 4 bytes
begin_time_seq4 = df_read[3]["temps"][0]
x_seq4 = df_read[3]["temps"] - begin_time_seq4
y_seq4 = df_read[3]["latence"]
y_seq4.tail()
```

```python
#plot des donn√©es des lectures seq par un pas de 4 bytes
plt.figure(figsize=(12, 3))
plt.plot(x_seq4[:500], y_seq4[:500], "-o",linewidth =2.5)
plt.xlabel("date d'arriv√©e(s)")
plt.ylabel("latence(s)")
plt.title('lecture de 16 Bytes par un pas de 4 Bytes')
plt.savefig("../outplot/read16B/read_step4.png")
plt.show()
```
