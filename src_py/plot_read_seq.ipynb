```python
import os 
from glob import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH = "../src/data_read_seq"
EXT = "*.csv"
```

```python
all_csv_files = [file
                 for path, subdir, files in os.walk(PATH)
                 for file in glob(os.path.join(path, EXT))]
```

```python
all_csv_files
```

```python
pd.options.display.float_format = '{:.6f}'.format
df_read = [pd.read_csv(file, delimiter=" ") for file in all_csv_files]
#df = pd.read_csv(all_csv_files[1], delimiter=",", names=["dates d'arrivée", "latence", "type","taille", "offset", "prio"])
```

```python
df_read[0].columns
```

```python
df_read[0]["latence"].describe()
```

```python
#données des lectures aléatoires de 4 bytes
begin_time_seq4 = df_read[0]["temps"][0]
x_seq4 = df_read[0]["temps"] - begin_time_seq4
y_seq4 = df_read[0]["latence"]
y_seq4.head()
```

```python
#plot données des lectures sequentielles
plt.figure(figsize=(12, 3))
plt.plot(x_seq4[0:500], y_seq4[0:500], "-o",linewidth =2.5)
plt.xlabel("date d'arrivée(s)")
plt.ylabel("latence(s)")
plt.title('lecture sequentielle de 4 bytes')
plt.savefig("../outplot/read/seq/read_seq4.png")
plt.show()
```
